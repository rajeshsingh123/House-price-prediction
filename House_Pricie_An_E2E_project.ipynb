TRANSFORMERS

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer


cat_col = ['ocean_proximity']
num_col = list(df.drop([*cat_col, 'median_house_value'], axis = 1).columns)

col_transf = ColumnTransformer([
                                ('onehot_cat', OneHotEncoder(), cat_col),
                                ('inputer', SimpleImputer(strategy = "median"), num_col),
                                ('num_scaling1', StandardScaler(), num_col),
])
```

Select and Train Model

```python
Y = df['median_house_value']
X = df.drop('median_house_value', axis = 1)
```

```python
X = col_transf.fit_transform(X)
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression

lin = LinearRegression()
lin.fit(X, Y)
```

```python
y_pred = lin.predict(X)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Y, y_pred)
mse
```

```python
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor()
tree.fit(X, Y)
```

```python
y_pred = tree.predict(X)

mse = mean_squared_error(Y, y_pred)
mse
```

```python
y_pred
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

xtrain, xval, ytrain, yval = train_test_split(X, Y, train_size=0.7)
#ask ques about size
```

```python
print('Linear Regression: ', mean_squared_error(lin_predict, yval))
print('Decision Tree    : ', mean_squared_error(tree_predict, yval))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(xtrain, ytrain)
forest_pred = forest.predict(xval)

print('Random Forest    : ', mean_squared_error(forest_pred, yval))
```

```python
print('Linear Regression: ', mean_squared_error(lin_predict, yval))
print('Decision Tree    : ', mean_squared_error(tree_predict, yval))
print('Random Forest    : ', mean_squared_error(forest_pred, yval))
```

Hyperparameter tuning

```python
forest = RandomForestRegressor(n_estimators = 50)
forest.fit(xtrain, ytrain)
forest_pred = forest.predict(xval)

print('Random Forest    : ', mean_squared_error(forest_pred, yval))
```

grid search CV

```python
from sklearn.model_selection import GridSearchCV

params = [
          {'n_estimators': [4, 6, 7], 'max_features': [2, 6]}, # 12
        ]

gscv = GridSearchCV(forest, params, scoring = 'neg_mean_squared_error', cv = 3, n_jobs=-1)#cv=3 {divedes data in 3 parts so that each part trains and makes the model
```

```python
gscv.fit(X, Y)
```

```python
gscv.cv_results_
```

```python
params = gscv.cv_results_['params']
scores = gscv.cv_results_['mean_test_score']

for res in zip(params, scores):
  print(res[0], -res[1])
```

```python
len(gscv.cv_results_['params'])
```

RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

params = {'n_estimators': [4, 6, 7, 10, 8, 9, 11, 12], 'max_features': [1, 2, 4, 5, 6, 7, 8]}

rscv = RandomizedSearchCV(forest, params, scoring = 'neg_mean_squared_error', cv = 3)

rscv.fit(X, Y)
# it randomly select from n estimators and max features for model
```

```python
rscv.cv_results_['params']
```

```python
params = rscv.cv_results_['params']
scores = rscv.cv_results_['mean_test_score']

for res in zip(params, scores):
  print(res[0], -res[1])
```
